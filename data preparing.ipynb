```python
import pandas as pd
project = pd.read_csv('/content/drive/MyDrive/Project/opendata_project.csv')
print(project.shape)
project.head()
```

#Data cleaning

##📑opendata_project table

###Select column

```python
project.columns
```

```python
project = project[['project_id', 'name_th', 'propertytype_id','propertytype_name_th', 'price_min','province_id','province_name_th']]
```

```python
project.head()
```

### Missing/Null Checking

```python
project.isnull().any()
```

```python
project.dtypes
```

#### ⇨ dropna

```python
# ' price_min ' ราคาต่ำสุดโครงการ
print(len(project[project['price_min'].isnull()])) #we used dropna()
project = project.dropna(subset=['price_min'])
```

```python
# 'province_id ' รหัสจังหวัด
print(len(project[project['province_id'].isnull()])) 
project = project.dropna(subset=['province_id'])
```

```python
# 'province_name_th ' ชื่อจังหวัด
print(len(project[project['province_name_th'].isnull()])) 
project = project.dropna(subset=['province_name_th'])
```

#### ⇨ fill Null with real world Knownlege

```python
project.isnull().any()
```

```python
#save data
project.to_csv('project_clean.csv', index=False)
```
