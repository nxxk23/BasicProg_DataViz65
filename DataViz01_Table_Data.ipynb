# Parallel Coordinates

````python
pd.plotting.parallel_coordinates(df, 'Name')
````

```setosa มีขนาดของกลีบเลี้ยงและกลีบดอกเล็กกว่าอีก 2 ชนิด และ ทั้ง2ชนิดได้แก่ versicolor และ virginica มีขนาดกลีบเลี้ยงและกลีบดอกยาวและลักษณะคล้ายกัน ```

## Subplot

````python
matplotlib.rcParams['figure.figsize']=[20,20]
````

````python
plt.subplot(1,3,1)  # subplot(มีกี่แถว, มีกี่หลัก, เป็นตัวที่เท่าไหร่)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest',cmap='hot') #data50ตัวแรก

plt.subplot(1,3,2)
plt.imshow(df.iloc[50:100,:-1],interpolation='nearest',cmap='hot')  #data 50 ตัวต่อไป

plt.subplot(1,3,3)
plt.imshow(df.iloc[100:150,:-1],interpolation='nearest',cmap='hot') #data 100 ถึงตัวที่ 150
````

````python
plt.subplot(1,4,2)
plt.imshow(df.iloc[50:100,:-1],interpolation='nearest',alpha = 0.8)

plt.subplot(1,4,1) #alpha คือ transparent
plt.imshow(df.iloc[:50,:-1],interpolation='nearest',cmap='hot')

plt.subplot(1,4,2)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest',cmap='hot',alpha = 0.6)

plt.subplot(1,4,3)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest',cmap='hot',alpha = 0.5)

plt.subplot(1,4,4)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest',cmap='hot',alpha = 0.8)
````
