#ตรวจสอบ Missing

```python
unit.isnull() #ตรวจสอบว่าค่านั้นเป็นค่า missing ไหม
```

```python
unit.isnull().any() # การเอาค่าความจริงใน column มา or กัน
# column ที่ False คือ ไม่มี missing
```

```python
unit.isnull().all() # การเอาค่าความจริงใน column มา and กัน ไม่ได้เกี่ยวกับค่า missing
# column ที่ true เป็นค่าว่างทั้งหมด
```

```python
unit.isnull().any().any()  # or มี 1 true = true
```

#ลบ missing

```python
unit.dropna() #count_elevator is all missing
```

```python
unit.dropna(axis=1) # drop column which contains missing values
```

```python
unit.dropna(how='all') 
```

```python
unit.iloc[0,:]
```

```python
unit.dropna(thresh=25)
```

```python
unit.dropna(axis=1, thresh=42000) # เก็บเฉพาะ column ที่มี missing อย่างน้อย int
```

```python
unit.head().dropna(axis=1, thresh=3)  #ตัด column ที่มีค่า missing อย่างน้อย 3 ตัว
```
