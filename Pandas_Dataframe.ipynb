2. พื้นที่และจำนวนห้องนอนของบ้านที่แพงที่สุด
3. จำนวนที่จอดรถและห้องน้ำของบ้านที่ถูกที่สุด
4. ชนิดของบ้านที่แพงที่สุด
5. บ้านชนิดไหนมีปริมาณมากที่สุด

```python
#1.ราคาเฉลี่ยของบ้านที่มีราคาระหว่าง 3-5 ล้านบาท คือ 497584.244124	
data1[data1['price_min']<3000000&5000000].describe()
```

```python
#2. พื้นที่และจำนวนห้องนอนของบ้านที่แพงที่สุด คือ NaN และ 3.0	
data1.describe() 
data1[data1['price_min']==580000000].describe()
```

```python
#3.จำนวนที่จอดรถและห้องน้ำของบ้านที่ถูกที่สุด คือ NaN	และ 1.0	
data1[data1['price_min']==100000].describe()
```

```python
#4. ชนิดของบ้านที่แพงที่สุด คือ Penthouse	
data1[data1['price_min']==580000000]
```

```python
#5. บ้านชนิดไหนมีปริมาณมากที่สุด คือ Warehouse / Factory
data1['propertytype_name_en'].max()
```
