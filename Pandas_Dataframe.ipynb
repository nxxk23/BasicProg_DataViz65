# Quiz  4

1. ราคาเฉลี่ยของบ้านที่มีราคาระหว่าง 3-5 ล้านบาท
2. พื้นที่และจำนวนห้องนอนของบ้านที่แพงที่สุด
3. จำนวนที่จอดรถและห้องน้ำของบ้านที่ถูกที่สุด
4. ชนิดของบ้านที่แพงที่สุด
5. บ้านชนิดไหนมีปริมาณมากที่สุด

```python
(data1['price_min']>3000000)
```

```python
(data1['price_min']<5000000)
```

```python
#1. ราคาเฉลี่ยของบ้านที่มีราคาระห่าง 3-5 ล้านบาท คือ 3,928,110
price_mean = data1[(data1['price_min']>3000000) & (data1['price_min']<5000000)]['price_min'].mean()
print(f'ราคาเฉลี่ยของบ้านที่มีราคาระห่าง 3-5 ล้านบาท คือ {int(price_mean):,}')
```

```python
#2. พื้นที่และจำนวนห้องนอนของบ้านที่แพงที่สุด คือ 1182.0 และ 3.0	
data1[data1['price_min']==580000000][['area_usable_min','count_room_bed']]
```

```python
data1[data1['price_min']==data1['price_min'].max()][['propertytype_name_en','area_usable_min','count_room_bed']] #เฉลย
```

```python
#3. จำนวนที่จอดรถและห้องน้ำของบ้านที่ถูกที่สุด คือ 
data1[data1['price_min']==data1['price_min'].min()][['count_parking','count_room_bath']] #เฉลย
```

```python
#4. ชนิดของบ้านที่แพงที่สุด คือ Condo
data1[data1['price_min']==580000000]
```

```python
#5. บ้านชนิดไหนมีปริมาณมากที่สุด คือ Detached House
data1['propertytype_name_en'].describe()
```
